# AppleStore data details
- 'id' : Application ID
- 'track_name' : Application name
- 'size_bytes' : Size (in Bytes)
- 'currency' : Application base currency used
- 'price' : Application price in appstore
- 'rating_count_tot' : User Rating counts (for all version)
- 'rating_count_ver' : User Rating counts (for current version)
- 'user_rating' : Average User Rating value (for all version)
- 'user_rating_ver' : Average User Rating value (for current version)
- 'ver' : Current application version
- 'cont_rating' : Content Rating
- 'prime_genre' : Main genre of application
- 'sup_devices.num' : Number of supporting devices
- 'ipadSc_urls.num' : Number of screenshots showed for display
- 'lang.num' : Number of supported languages
- 'vpp_lic' : Vpp Device Based Licensing Enabled
- 'app_desc' : Application description

## Creating a Model for Rating : Attempt 1


```python
# Import the encoder from sklearn
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

# OneHotEncoding of categorical predictors (not the response)
applestore_cat = applestore[['cont_rating','prime_genre']]
ohe.fit(applestore_cat)
applestore_cat_ohe = pd.DataFrame(ohe.transform(applestore_cat).toarray(), 
                                  columns=ohe.get_feature_names_out(applestore_cat.columns))

# Check the encoded variables
applestore_cat_ohe.info()
```

```python
# Combining Numeric features with the OHE Categorical features
# 'size_bytes','rating_count_tot','sup_devices.num','ipadSc_urls.num','lang.num','vpp_lic'
applestore_num = applestore[['size_bytes','rating_count_tot','sup_devices.num','ipadSc_urls.num','lang.num','vpp_lic']]
applestore_res = applestore['goodapp']
applestore_ohe = pd.concat([applestore_num, applestore_cat_ohe, applestore_res], 
                           sort = False, axis = 1).reindex(index=applestore_num.index)

# Check the final dataframe
applestore_ohe.info()
```

```python
# Import essential models and functions from sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import plot_tree

# Extract Response and Predictors
y = pd.DataFrame(applestore_ohe['goodapp'])
X = pd.DataFrame(applestore_ohe.drop('goodapp', axis = 1))

# Split the Dataset into Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

# Decision Tree using Train Data
dectree = DecisionTreeClassifier(max_depth = 4)  # change max_depth to experiment
dectree.fit(X_train, y_train)                    # train the decision tree model

# Plot the trained Decision Tree
f = plt.figure(figsize=(24,24))
plot_tree(dectree, filled=True, rounded=True, 
          feature_names=X_train.columns, 
          class_names=["True","False"])
```

## Training Results

```python
# Predict the Response corresponding to Predictors
y_train_pred = dectree.predict(X_train)

# Print the Classification Accuracy
print("Train Data")
print("Accuracy  :\t", dectree.score(X_train, y_train))
print()

# Print the Accuracy Measures from the Confusion Matrix
cmTrain = confusion_matrix(y_train, y_train_pred)
tpTrain = cmTrain[1][1] # True Positives : Good (1) predicted Good (1)
fpTrain = cmTrain[0][1] # False Positives : Bad (0) predicted Good (1)
tnTrain = cmTrain[0][0] # True Negatives : Bad (0) predicted Bad (0)
fnTrain = cmTrain[1][0] # False Negatives : Good (1) predicted Bad (0)

print("TPR Train :\t", (tpTrain/(tpTrain + fnTrain)))
print("TNR Train :\t", (tnTrain/(tnTrain + fpTrain)))
print()

print("FPR Train :\t", (fpTrain/(tnTrain + fpTrain)))
print("FNR Train :\t", (fnTrain/(tpTrain + fnTrain)))

# Plot the two-way Confusion Matrix
sns.heatmap(confusion_matrix(y_train, y_train_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18})
```

## Test Results

```python
# Import the required metric from sklearn
from sklearn.metrics import confusion_matrix

# Predict the Response corresponding to Predictors
y_test_pred = dectree.predict(X_test)

# Print the Classification Accuracy
print("Test Data")
print("Accuracy  :\t", dectree.score(X_test, y_test))
print()

# Print the Accuracy Measures from the Confusion Matrix
cmTest = confusion_matrix(y_test, y_test_pred)
tpTest = cmTest[1][1] # True Positives : Good (1) predicted Good (1)
fpTest = cmTest[0][1] # False Positives : Bad (0) predicted Good (1)
tnTest = cmTest[0][0] # True Negatives : Bad (0) predicted Bad (0)
fnTest = cmTest[1][0] # False Negatives : Good (1) predicted Bad (0)

print("TPR Test :\t", (tpTest/(tpTest + fnTest)))
print("TNR Test :\t", (tnTest/(tnTest + fpTest)))
print()

print("FPR Test :\t", (fpTest/(fpTest + tnTest)))
print("FNR Test :\t", (fnTest/(fnTest + tpTest)))

# Plot the two-way Confusion Matrix
sns.heatmap(confusion_matrix(y_test, y_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18})
```

## Create a Model for Rating : Attempt 2

Let's refine the previous model by balancing the classes of the response `Rating` in the training data. Keeping the OneHotEncoding same as before.

```python
applestore_ohe.head()
```

```python
# Upsample Bad to match Good
from sklearn.utils import resample

applestoreGood = applestore_ohe[applestore_ohe.goodapp == True]
applestoreBad = applestore_ohe[applestore_ohe.goodapp == False]
 
# Upsample the Bad samples
applestoreBad_up = resample(applestoreBad, 
                        replace=True,                     # sample with replacement
                        n_samples=applestoreGood.shape[0])    # to match number of Good
 
# Combine the two classes back after upsampling
applestore_ohe_up = pd.concat([applestoreGood, applestoreBad_up])
 
# Check the ratio of the classes
applestore_ohe_up['goodapp'].value_counts()
```

```python
# Quick plot to check the balanced classes visually
sns.catplot(y = 'goodapp', data = applestore_ohe_up, kind = "count")
```

## Creating a Model for Rating : Attempt 4

Let's try out a basic set of tools for Cross-Validation, so that we can tune the Hyper-parameters of Random Forest in this case.

```python
# Import RandomForestClassifier model from Scikit-Learn
from sklearn.ensemble import RandomForestClassifier

# Extract Response and Predictors
y = pd.DataFrame(applestore_ohe_up['goodapp'])
X = pd.DataFrame(applestore_ohe_up.drop('goodapp', axis = 1))

# Split the Dataset into Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# Import GridSearch for hyperparameter tuning using Cross-Validation (CV)
from sklearn.model_selection import GridSearchCV

# Define the Hyper-parameter Grid to search on, in case of Random Forest
param_grid = {'n_estimators': np.arange(100,1001,100),   # number of trees 100, 200, ..., 1000
              'max_depth': np.arange(2, 11)}             # depth of trees 2, 3, 4, 5, ..., 10

# Create the Hyper-parameter Grid
hpGrid = GridSearchCV(RandomForestClassifier(),   # the model family
                      param_grid,                 # the search grid
                      cv = 5,                     # 5-fold cross-validation
                      scoring = 'accuracy')       # score to evaluate

# Train the models using Cross-Validation
hpGrid.fit(X_train, y_train.goodapp.ravel())
```

```python
# Fetch the best Model or the best set of Hyper-parameters
print(hpGrid.best_estimator_)

# Print the score (accuracy) of the best Model after CV
print(np.abs(hpGrid.best_score_))
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from sklearn.model_selection import train_test_split
import lightgbm
from sklearn.metrics import roc_auc_score

import matplotlib.pylab as plt


def test(row):  
    if row['goodapp'] == True:
        return 1
    return 0

applestore_ohe_up['goodapp'] = applestore_ohe_up.apply(lambda row: test(row), axis=1)

# Extract Response and Predictors
y = pd.DataFrame(applestore_ohe_up['goodapp'])
X = pd.DataFrame(applestore_ohe_up.drop('goodapp', axis = 1))

# Split the Dataset into Train and Test
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
train_data = lightgbm.Dataset(X_train, label=y_train)
valid_data = lightgbm.Dataset(X_valid, label=y_valid)
```

```python
parameters = {'objective': 'binary',
              'metric': 'auc',
              'is_unbalance': 'true',
              'boosting': 'gbdt',
              'num_leaves': 63,
              'feature_fraction': 0.5,
              'bagging_fraction': 0.5,
              'bagging_freq': 20,
              'learning_rate': 0.01,
              'verbose': -1
             }
```

```python
model_lgbm = lightgbm.train(parameters,
                            train_data,
                            valid_sets=valid_data,
                            num_boost_round=5000, #n_estimators
                            early_stopping_rounds=50)
```

```python
y_train_pred = model_lgbm.predict(X_train)
y_valid_pred = model_lgbm.predict(X_valid)

print(" Train accurancy: {:.4f}\n Test accurancy: {:.4f}".format(roc_auc_score(y_train, y_train_pred),
                                                    roc_auc_score(y_valid, y_valid_pred)))
```
